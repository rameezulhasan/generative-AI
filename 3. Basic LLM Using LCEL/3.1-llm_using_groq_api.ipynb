```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
model = ChatGroq(model="openai/gpt-oss-120b", groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="Translate the following from English to urdu"),
    HumanMessage(content="Hello, How are you?")
]

result = model.invoke(messages)
result
```

```python
#parser
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
chain = model|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
generic_template = "Translate into following language {language}"

prompt = ChatPromptTemplate.from_messages(
    [
        ("system",generic_template),
        ("user","{text}")
    ]
)

prompt_generated = prompt.invoke({"language":"French",
               "text":"Where are you going?"})
prompt_generated
```

```python
prompt_generated.to_messages()
```

```python
chain = prompt|model|parser
chain.invoke({"language":"Roman English",
               "text":"Where are you going?"})
```
