## StuffDocumentChain Text summarization

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")
text = loader.load_and_split()
text
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system","Write a concise and short summary of the following speech"),
        ("human","""Speech:{text}""")
    ]
)
```

```python
chain = prompt | llm
```

```python
context = "\n\n".join(doc.page_content for doc in text)
summary = chain.invoke({"text": context})
```

```python
summary
```

### Map reduce to Summarize Large documents

```python
from langchain_classic.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500,chunk_overlap=100).split_documents(text)
text_splitter
```

```python
map_prompt = ChatPromptTemplate.from_template("""
Write a concise summary of the following text.

{text}
""")
```

```python
map_chain = map_prompt | llm | StrOutputParser()
```

```python
summaries = []

for chunk in text_splitter:
    summary = map_chain.invoke(
        {
            "text":chunk.page_content
        }
    )
    summaries.append(summary)
```

```python
summaries
```

```python
combined_summary = "\n\n".join(summaries)
```

```python
combined_summary
```

```python
reduced_prompt = ChatPromptTemplate.from_template(
    """
These are summaries of different parts of a document
{summaries}
create one final summary.
"""
)
reduced_chain = reduced_prompt | llm | StrOutputParser()
```

```python
final_summary = reduced_chain.invoke(
    {
        "summaries":combined_summary
    }
)
```

```python
final_summary
```

```python
tokens = llm.get_num_tokens(combined_summary)

print(tokens)
```

```python
tokens = llm.get_num_tokens(final_summary)

print(tokens)
```
