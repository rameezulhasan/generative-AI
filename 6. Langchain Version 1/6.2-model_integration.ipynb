## Streaming and Batch

```python
model.invoke("write a 1000 words blog on How to earn money online with facebook in 2026. blog should be seo friendly, include headings, subheadings, and bullet points. blog should be in english language. and write blog as you are expert content writer and digital marketing expert. blog should be unique and plagiarism free. blog should be in conversational tone. blog should be in simple language. blog should be easy to read and understand. blog should be engaging and informative. blog should be actionable and practical. blog should be well-structured and organized. blog should be visually appealing and attractive. blog should be optimized for search engines. blog should be shareable and linkable. blog should be mobile-friendly and responsive. blog should be fast-loading and lightweight. blog should be secure and safe. blog should be accessible and inclusive. blog should be compliant with laws and regulations. blog should be ethical and responsible. blog should be transparent and honest. blog should be authentic and genuine. blog should be credible and trustworthy. blog should be authoritative and influential. blog should be relevant and timely. blog should be valuable and useful. blog should be interesting and entertaining. blog should be inspiring and motivating. blog should be educational and informative. blog should be thought-provoking and stimulating. blog should be reflective and introspective. blog should be analytical and critical. blog should be creative and innovative. blog should be original and unique.")
```

```python
for chunk in model.stream("write a 1000 words blog on How to earn money online with facebook in 2026"):
    print(chunk.text, end="|", flush=True)
```

```python
for chunk in model.stream("why parrots are so intelligent"):
    print(chunk.text, end="", flush=True)
```

## BATCH

```python
responses = model.batch([
    "why parrots are so intelligent",
    "how do airplanes fly",
    "what is the capital of France"
])

for response in responses:
    print(response)
```
