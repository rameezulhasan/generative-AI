```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
groq_api_key = os.getenv("GROQ_API_KEY")
model = ChatGroq(model="openai/gpt-oss-120b", groq_api_key=groq_api_key)
```

```python
from langchain_core.messages import HumanMessage
prompt = [
    HumanMessage(content="Hi my name is Rameez")
]
response = model.invoke(prompt)
response.content
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(response)
```

```python
chain = model|parser
chain.invoke(prompt)
```

```python
from langchain_core.messages import HumanMessage
prompt = [
    HumanMessage(content="Hi what's my name?")
]
response = model.invoke(prompt)
response.content
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}

def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model, get_session_history)
```

```python
config1 = {"configurable":{"session_id":"chat1"}}
```

```python
response = with_message_history.invoke(
    [HumanMessage(content="Hi My name is Rameez Ul Hassan and I like to play Cricket")],
    config=config1
)
response.content
```

```python
response = with_message_history.invoke(
    [HumanMessage(content="what is my name? and what i like")],
    config=config1
)
response.content
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
prompt_templates = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant. Answer all the questions to best of your ability."),
        MessagesPlaceholder(variable_name="messages")
    ]
)

chain = prompt_templates|model
```

```python
chain.invoke({"messages":[HumanMessage(content="Hi what is my name ?")]})
```

```python
with_message_history = RunnableWithMessageHistory(chain, get_session_history)
```

```python
config3 = {"configurable":{"session_id":"chat3"}}
```

```python
response = with_message_history.invoke(
    [
        HumanMessage(content="Tell me which programming language is most important for Artificiall Intelligence?")
    ], config=config3
)

response.content
```

```python
response = with_message_history.invoke(
    [
        HumanMessage(content="Which question i asked in previous chat?")
    ], config=config3
)

response.content
```
